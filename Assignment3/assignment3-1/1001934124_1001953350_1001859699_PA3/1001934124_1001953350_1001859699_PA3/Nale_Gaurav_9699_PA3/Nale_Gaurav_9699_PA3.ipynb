```python
print ('# # # # # # # # #  Code for Task 3, Case:1 # # # # # # # # #') 
transaction_encoder = TransactionEncoder()
transaction_encoder_array = transaction_encoder.fit(dataToList).transform(dataToList)
df = pd.DataFrame(transaction_encoder_array, columns=transaction_encoder.columns_)
frequent_itemsets=fpgrowth(df, min_support=0.01, use_colnames=True)
print ('Case 1 (minimum support=0.01)')
print ('Case 1 Reasoning: FPGrowth provided result faster than apriori. We get maximum confidence in following rule ("weekday", "soup") ==> ("afternoon") , Confidence: 0.951 . Its clear that with a lower support threshold we can create more rules, but the confidence of the rules is very low. Thus, if people purchase "soup" in "weekday" together then there is 95.1% chance that they will purchase in "afternoon".')
print ('Case 1 Output:')
res=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
print('***************Printing all rules:*****************')
print(res)
print('*****************Getting highest confidence*****************')
res1=res[['antecedents','consequents','support','confidence','lift']]
print("Returning the items and itemsets with at least 85% confidence:")
res1[res1['confidence']>=0.85]
```

```python
print ('# # # # # # # # #  Code for Task 3, Case:2 # # # # # # # # #') 
print ('Case 2 (minimum support=0.015)')
print ('Case 2 Reasoning:FPGrowth provided result faster than apriori. We get less rules with high threshold values, in following rule ("Soup") ==> ("afternoon") , Confidence: 0.947 . Its clear that with a higher support threshold, can create less rules, but the confidence of the rules increases. Thus, if people purchase "soup" then there is 94.7% chance that they will purchase in "afternoon".')
print ('Case 2 Output:')
frequent_itemsets=fpgrowth(df, min_support=0.015, use_colnames=True)
res=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.15)
print('***************Printing all rules:*****************')
print(res)
print('*****************Getting highest confidence*****************')
res1=res[['antecedents','consequents','support','confidence','lift']]
print("Returning the items and itemsets with at least 85% confidence:")
res1[res1['confidence']>=0.85]
```

```python
print ('# # # # # # # # #  Code for Task 3, Case:3 # # # # # # # # #') 
print ('Case 3 (minimum support=0.04)')
print ('Case 3 Reasoning:FPGrowth provided result faster than apriori. We get maximum confidence in following rule ("Tea") ==> ("Weekday") , Confidence: 0.678 . Its clear that with a lower support threshold we can create more rules, but the confidence of the rules is very low. Thus, there is 67.8%  chance that people will have "tea" in "weekday"')
print ('Case 3 Output:')
frequent_itemsets=fpgrowth(df, min_support=0.04, use_colnames=True)

res = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
print('***************Printing all rules:*****************')
print(res)
print('*****************Getting highest confidence*****************')
res1=res[['antecedents','consequents','support','confidence','lift']]
print("Returning the items and itemsets with at least 50% confidence:")
res1[res1['confidence']>=0.5]
```
