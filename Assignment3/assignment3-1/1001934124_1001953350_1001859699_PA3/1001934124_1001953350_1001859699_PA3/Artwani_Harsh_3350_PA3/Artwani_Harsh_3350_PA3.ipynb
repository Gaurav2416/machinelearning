```python
print ('# # # # # # # # #  Code for Task 3, Case:1 # # # # # # # # #') 
transaction_encoder = TransactionEncoder()
transaction_encoder_array = transaction_encoder.fit(dataToList).transform(dataToList)
df = pd.DataFrame(transaction_encoder_array, columns=transaction_encoder.columns_)
frequent_itemsets=fpgrowth(df, min_support=0.0001, use_colnames=True)
print ('Case 1 (minimum support=0.0001)')
print ('Case 1 Reasoning: At a glance, there are many rules with confidence = 1. We get many rules for minimum support of 0.0001 and minimum threshold of 0.0015. More rules means better association of items and their chance of reccurence. Here we have a maximum support for id 1669 and date 19-08-2015 combination(s = 0.000232 and confidence = 1).')
print ('Case 1 Output:')
res=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.0015)
print('***************Printing all rules:*****************')
print(res)
print('*****************Getting highest confidence*****************')
res1=res[['antecedents','consequents','support','confidence','lift']]
print("Returning the items and itemsets with equal to 100% confidence:")
res1[res1['confidence']==1] 
```

```python
print ('# # # # # # # # #  Code for Task 3, Case:2 # # # # # # # # #') 
print ('Case 2 (minimum support=0.01)')
print ('Case 2 Reasoning: At a glance, there are many rules with confidence = 1. We get many rules for minimum support of 0.0001 and minimum threshold of 0.0015. More rules means better association of items and their chance of reccurence. Here we have a maximum support for id 1669 and date 19-08-2015 combination(s = 0.000232 and confidence = 1).')
print ('Case 2 Output:')
frequent_itemsets=fpgrowth(df, min_support=0.000025, use_colnames=True)
res=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
print('***************Printing all rules:*****************')
print(res)
print('*****************Getting highest confidence*****************')
res1=res[['antecedents','consequents','support','confidence','lift']]
print("Returning the items and itemsets with confidence: 100% and support:0.02%")
res2= res1[res1['support']>=0.0002]
res2[res2['confidence']==1]
```

```python
print ('# # # # # # # # #  Code for Task 3, Case:3 # # # # # # # # #') 
print ('Case 3 (minimum support=0.001)')
print ('Case 3 Reasoning: We get very less rules for minimum support of 0.00005 and but with very strong confidence of 1.0 at min confidence threshhold 0.07. Here we have a maximum support of 0.000232 with confidence 1.Rule is with antecedent 1669 and consequent as 19-08-2015')
print ('Case 3 Output:')
frequent_itemsets=fpgrowth(df, min_support=0.00005, use_colnames=True)
res = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.07)
print('***************Printing all rules:*****************')
print(res)
print('*****************Getting highest confidence*****************')
res1=res[['antecedents','consequents','support','confidence','lift']]
print("Returning the items and itemsets with confidence: 100% and support:0.02%")
res2= res1[res1['support']>=0.0002]
res2[res2['confidence']==1]
```
