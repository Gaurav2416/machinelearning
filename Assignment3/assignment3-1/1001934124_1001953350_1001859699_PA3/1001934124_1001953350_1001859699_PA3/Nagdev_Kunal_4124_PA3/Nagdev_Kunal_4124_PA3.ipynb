```python
print ('# # # # # # # # #  Code for Task 3, Case:1 # # # # # # # # #') 
transaction_encoder = TransactionEncoder()
transaction_encoder_array = transaction_encoder.fit(dataToList).transform(dataToList)
df = pd.DataFrame(transaction_encoder_array, columns=transaction_encoder.columns_)

print("Returning the items and itemsets with at least 1% support:")
frequent_itemsets=fpgrowth(df, min_support=0.01, use_colnames=True)
print ('Case 1 (minimum support=0.01)')
print ('Case 1 Reasoning:  FPGrowth is faster and accurate than apriori. We get maximum confidance in following rule ("eggs", "ground beef") ==> ("mineral water") , Confidence: 0.506 . Its clear that with a lower support threshold we can create more rules, but the confidence of the rules is very low. Thus, if people purchase "eggs and groundbeef" together then there is 50.6% chance that they will also going to purchase "mineral water" along with those items.')
print ('Case 1 Output:')
res=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
print('***************Printing all rules:*****************')
print(res)
print('*****************Getting highest confidence*****************')
res1=res[['antecedents','consequents','support','confidence','lift']]
print("Returning the items and itemsets with at least 45% confidence:")
res1[res1['confidence']>=0.45]
```

```python
print ('# # # # # # # # #  Code for Task 3, Case:2 # # # # # # # # #') 
print ('Case 2 (minimum support=0.015)')
print ('Case 2 Reasoning: FPGrowth is faster and accurate than apriori. We get moderate confidance in following rule ("soup", ) ==> ("mineral water") , Confidence: 0.456 . Its clear that increasing threshhold value creates lower rules, but the confidence of the rules is more. Thus, if people purchase "soup" together then there is 45.6% chance that they will also going to purchase "mineral water" along with those items.')
print ('Case 2 Output:')

print("Returning the items and itemsets with at least 0.05% support:")
frequent_itemsets=fpgrowth(df, min_support=0.015, use_colnames=True)

res=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.15)
print('***************Printing all rules:*****************')
print(res)
print('*****************Getting highest confidence*****************')
res1=res[['antecedents','consequents','support','confidence','lift']]
print("Returning the items and itemsets with at least 40% confidence:")
res1[res1['confidence']>=0.40]
```

```python
print ('# # # # # # # # #  Code for Task 3, Case:3 # # # # # # # # #') 
print ('Case 3 (minimum support=0.04)')
print ('Case 3 Reasoning: FPGrowth is faster and accurate than apriori.Confidence value further decreases when we increase the support, also increasing the confidence will reduce the total number of rules generated. Thus, if people purchase "ground beef" together then there is 41.7% chance that they will also going to purchase "mineral water" along with those items. In comparision to previous support values, there is total decrease of 10% in final values.')
print ('Case 3 Output:')


frequent_itemsets=fpgrowth(df, min_support=0.04, use_colnames=True)

res = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
print('***************Printing all rules:*****************')
print(res)
print('*****************Getting highest confidence*****************')
res1=res[['antecedents','consequents','support','confidence','lift']]
print("Returning the items and itemsets with at least 35% confidence:")
res1[res1['confidence']>=0.35]
```
